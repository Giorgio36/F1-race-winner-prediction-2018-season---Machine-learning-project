```python
raw_data['Driver Name'] = raw_data['forename'] + ' ' + raw_data['surname']
raw_data['date'] = pd.to_datetime(raw_data['date'])
raw_data['dob'] = pd.to_datetime(raw_data['dob'])
raw_data = raw_data.drop(['forename','surname'],axis=1)
```

```python
raw_data = raw_data.rename(columns={"name_x":"Grand Prix Name", "name_y":"Constructor Name", "grid":"Grid Position", "positionText":"Race Position",
                         "nationality_x": "Driver Nationality", "nationality_y": "Constructor Nationality", "country" : "Country", "statusId" : "DNF"})
```

```python
#Taking into account the data from the last 40 years until 2018. This, in order to test the algorithms in the 2019 season, which was one with several different
#winners and could be a bit more challenging to get right
dataset = raw_data[(raw_data['year'] > 1984) & (raw_data['year'] < 2024)]
dataset = dataset[['year', 'round', 'raceId', 'circuitId', 'Country', 'Grand Prix Name', 'date','constructorId', 'Constructor Name',
             'driverId','Driver Name','Grid Position', 'Race Position', 'DNF', 'dob',
             'Driver Nationality', 'Constructor Nationality']]
```

```python
dataset['year'].unique()
```

```python
#Mapping teams's old names to current names
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Manor Marussia' if (x=='Sauber' or x== 'Marussia') else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Mercedes' if (x=='Brawn' or x== 'Honda' or x== 'BAR') else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Aston Martin' if (x=='Force India' or x== 'Racing Point' or
                                                                                             x== 'Spyker' or x== 'Spyker MF1') else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Alpha Tauri' if (x=='Toro Rosso' or x== 'AlphaTauri') else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Alpine' if (x=='Renault' or x== 'Lotus F1' or
                                                                                             x== 'Spyker' or x== 'Benetton') else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Red Bull' if x=='Jaguar' else x)
dataset['Constructor Name'] = dataset['Constructor Name'].apply(lambda x: 'Alfa Romeo' if (x=='Sauber' or x== 'BMW Sauber') else x)

dataset['Constructor Name'].unique()
```

```python
#Taking into account that all the letters refer to a driver who did not fisnish the race, all theses cases will be considered as such and will be mapped to
#position number zero later, to work with numerical data.
dataset['Race Position'] = dataset['Race Position'].apply(lambda x: 'R' if ( x=='W' or x=='D' or x=='E' or x=='F' or x=='N') else x)
dataset['Race Position'] = dataset['Race Position'].apply(lambda x: '0' if x =='R' else x)
dataset['Race Position'].unique()
```

```python
dataset[['Grid Position', 'Race Position','DNF']] = dataset[['Grid Position', 'Race Position','DNF']].apply(pd.to_numeric)
dataset[['Grand Prix Name','Constructor Name', 'Driver Name', 'Driver Nationality', 'Constructor Nationality']] = dataset[['Grand Prix Name',
                                                              'Constructor Name', 'Driver Name', 'Driver Nationality', 'Constructor Nationality']].astype(str)
dataset.dtypes
```

```python
#Inserting driver and constructor win columns
dataset['Driver Win'] = dataset['Race Position'].apply(lambda x: 1 if x == 1 else 0)
dataset['Constructor Win'] = dataset['Race Position'].apply(lambda x: 1 if x == 1 else 0)
dataset.sort_values(by=['year'],inplace=True)
```

```python
#Inserting home race for driver and constructor
dataset['Country'] = dataset['Country'].apply(lambda x: x[:3])
dataset['Driver Nationality'] = dataset['Driver Nationality'].apply(lambda x: x[:3])
dataset['Constructor Nationality'] = dataset['Constructor Nationality'].apply(lambda x: x[:3])
dataset['Driver Nationality'] = dataset['Driver Nationality'].apply(lambda x: 'Fre' if x == 'Fra' else x)
dataset['Driver Nationality'] = dataset['Driver Nationality'].apply(lambda x: 'USA' if x == 'Ame' else x)
dataset['Driver Nationality'] = dataset['Driver Nationality'].apply(lambda x: 'UK' if x == 'Bri' else x)
dataset['Constructor Nationality'] = dataset['Constructor Nationality'].apply(lambda x: 'Fre' if x == 'Fra' else x)
dataset['Constructor Nationality'] = dataset['Constructor Nationality'].apply(lambda x: 'USA' if x == 'Ame' else x)
dataset['Constructor Nationality'] = dataset['Constructor Nationality'].apply(lambda x: 'UK' if x == 'Bri' else x)
dataset['Driver Home Race'] = dataset['Driver Nationality']==dataset['Country']
dataset['Constructor Home Race'] = dataset['Constructor Nationality']==dataset['Country']
dataset[['Driver Home Race','Constructor Home Race']] = dataset[['Driver Home Race','Constructor Home Race']].astype(int)
dataset['DNF (Constructor Fault)'] = dataset['DNF'].apply(lambda x: 1 if (x not in [3,4,20] and x not in [1,11,12,13,14,15,16,17]) else 0)
dataset['DNF (Driver Error)'] = dataset['DNF'].apply(lambda x: 1 if x in [3,4,20] else 0)
dataset['DNF'] = dataset['DNF'].apply(lambda x: 1 if x not in [1,11,12,13,14,15,16,17] else 0)
```

```python
dataset[(dataset['year']==2015) & (dataset['Driver Name'] == 'Fernando Alonso')]
```

```python
#Getting the DNF ratio for every driver and constructor during each season to avoid data leakage. Additionally,for each season the data from the previous year would be taking into account in
#the dataset, not the data for the same year, avoiding this way looking into the future and data leakage.

#Grouping by driver
dnf_driver = dataset.groupby(['year','Driver Name'])['DNF (Driver Error)'].sum()
races_num_driver = dataset.groupby(['year','Driver Name'])['DNF (Driver Error)'].count()
dnf_norm_driver = dnf_driver/races_num_driver
```

```python
dnf_ratio_driver = pd.DataFrame()
dnf_ratio_driver['year'] = ""
dnf_ratio_driver['Driver Name'] = ""
dnf_ratio_driver['DNF Ratio (Driver Error)'] = ""

dnf_year_driver = []
dnf_driver = []
dnf_val_driver = []

for i in range(len(dnf_norm_driver.index)):
  dnf_year_driver.append(dnf_norm_driver.index[i][0])
  dnf_driver.append(dnf_norm_driver.index[i][1])
  dnf_val_driver.append(dnf_norm_driver[dnf_norm_driver.index[i][0]][dnf_norm_driver.index[i][1]])
```

```python
dnf_ratio_driver['year'] = dnf_year_driver
dnf_ratio_driver['Driver Name'] = dnf_driver
dnf_ratio_driver['DNF Ratio (Driver Error)'] = dnf_val_driver
```

```python
#Grouping by constructor
dnf_constructor = dataset.groupby(['year','Constructor Name'])['DNF (Constructor Fault)'].sum()
races_num_constructor = dataset.groupby(['year','Constructor Name'])['DNF (Constructor Fault)'].count()
dnf_norm_constructor = dnf_constructor/races_num_constructor
```

```python
dnf_ratio_constructor = pd.DataFrame()
dnf_ratio_constructor['year'] = ""
dnf_ratio_constructor['Constructor Name'] = ""
dnf_ratio_constructor['DNF Ratio (Constructor Fault)'] = ""

dnf_year_constructor = []
dnf_constructor = []
dnf_val_constructor = []

for i in range(len(dnf_norm_constructor.index)):
  dnf_year_constructor.append(dnf_norm_constructor.index[i][0])
  dnf_constructor.append(dnf_norm_constructor.index[i][1])
  dnf_val_constructor.append(dnf_norm_constructor[dnf_norm_constructor.index[i][0]][dnf_norm_constructor.index[i][1]])
```

```python
dnf_ratio_constructor['year'] = dnf_year_constructor
dnf_ratio_constructor['Constructor Name'] =  dnf_constructor
dnf_ratio_constructor['DNF Ratio (Constructor Fault)'] =  dnf_val_constructor
```
